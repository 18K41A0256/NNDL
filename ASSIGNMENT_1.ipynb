```
from sympy import *
p=symbols('p')
f=p*4+3*p*2+10
q=diff(f,p) 
print(q)
P=int(input("enter p value:"))
itr=1 
n=0.1 
epoch=2
while (itr <= epoch) :
   Q=q.subs(p,P)
   print("q",Q)
   deltap=-n*Q
   print("delta(p) =",deltap)
   newp = P + deltap
   print("p =",newp)
   itr= itr + 1
   P=newp
print("final p =",P)
print("Minimum value is",f.subs(p,P))
```
